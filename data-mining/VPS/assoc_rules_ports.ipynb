```python
import pandas as pd

df = pd.read_json("../../datas/vps/detected.json", lines=True)

df.head(5)
```

```python
ports_matrix = []
tmp = []
sess_uuids_ports = {
  # "{sess_uuid}": []
}
sess_uuids = []

for _, row in df.iterrows():
  sess_uuid = row.get("cookies", {}).get("sess_uuid") or row.get("response_msg", {}).get("response", {}).get("message", {}).get("sess_uuid")
  port = row["peer"]["port"]

  if sess_uuid in sess_uuids:
    sess_uuids_ports[sess_uuid].append(port)
  else:
    sess_uuids_ports[sess_uuid] = [port]
    sess_uuids.append(sess_uuid)

ports_matrix = list(sess_uuids_ports.values())
# for paths in sess_uuids_ports.values():
#   ports_matrix.append(paths)

sess_uuids_ports
# ports_matrix
# print(len(ports_matrix), len(sess_uuids_ports.keys()))
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth

te = TransactionEncoder()
te_ary = te.fit(ports_matrix).transform(ports_matrix)
df = pd.DataFrame(te_ary, columns=te.columns_)

df
```

```python
# frequent_itemsets = fpgrowth(df, min_support=0.3, use_colnames=True)
frequent_itemsets = fpgrowth(df, min_support=0.0038, use_colnames=True)
### alternatively:
#frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
#frequent_itemsets = fpmax(df, min_support=0.6, use_colnames=True)

frequent_itemsets
```

```python
frequent_itemsets["length"] = frequent_itemsets["itemsets"].apply(lambda x: len(x))
### alternatively:
#frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
#frequent_itemsets = fpmax(df, min_support=0.6, use_colnames=True)

frequent_itemsets
```

```python
large_frequent_itemsets = frequent_itemsets[frequent_itemsets["length"] >= 3]
large_frequent_itemsets
```

```python
import psycopg2

conn = psycopg2.connect(database="web_honeypot_vps_new", user = "postgres", password = "admin", host = "127.0.0.1", port = "5432")

print("Opened database successfully")
```

```python
# create table
cur = conn.cursor()
cur.execute('''CREATE TABLE assoc_rules_ports (
            ID INT PRIMARY KEY     NOT NULL,
            SUPPORT           REAL    NOT NULL,
            PORT            VARCHAR(255)     NOT NULL);''')

print("Table created successfully")

conn.commit()
```

```python
# insert data

cur = conn.cursor()

insert_query = """
    INSERT INTO assoc_rules_ports (ID, SUPPORT, PORT)
    VALUES (%s, %s, %s)
"""

for idx, row in frequent_itemsets.iterrows():
    cur.execute(
        insert_query,
        (int(idx), float(row["support"]), str(list(row["itemsets"])))
    )

conn.commit()
print("Records created successfully")
```
