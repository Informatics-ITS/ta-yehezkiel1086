```python
import pandas as pd

df = pd.read_json("../../datas/vps/detected.json", lines=True)

df.head(5)
```

```python
ips_matrix = []
tmp = []
sess_uuids_ips = {
  # "{sess_uuid}": []
}
sess_uuids = []

for _, row in df.iterrows():
  sess_uuid = row.get("cookies", {}).get("sess_uuid") or row.get("response_msg", {}).get("response", {}).get("message", {}).get("sess_uuid")
  ip = row["peer"]["ip"]

  if sess_uuid in sess_uuids:
    sess_uuids_ips[sess_uuid].append(ip)
  else:
    sess_uuids_ips[sess_uuid] = [ip]
    sess_uuids.append(sess_uuid)

ips_matrix = list(sess_uuids_ips.values())
# for paths in sess_uuids_ips.values():
#   ips_matrix.append(paths)

sess_uuids_ips
# ips_matrix
# print(len(ips_matrix), len(sess_uuids_paths.keys()))
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth

te = TransactionEncoder()
te_ary = te.fit(ips_matrix).transform(ips_matrix)
df = pd.DataFrame(te_ary, columns=te.columns_)

df
```

```python
# frequent_itemsets = fpgrowth(df, min_support=0.3, use_colnames=True)
frequent_itemsets = fpgrowth(df, min_support=0.00001, use_colnames=True)
### alternatively:
#frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
#frequent_itemsets = fpmax(df, min_support=0.6, use_colnames=True)

frequent_itemsets.head(20)
```

```python
import psycopg2

conn = psycopg2.connect(database="web_honeypot_vps_new", user = "postgres", password = "admin", host = "127.0.0.1", port = "5432")

print("Opened database successfully")
```

```python
# create table
cur = conn.cursor()
cur.execute('''CREATE TABLE assoc_rules_ips (
            ID INT PRIMARY KEY     NOT NULL,
            SUPPORT           REAL    NOT NULL,
            IP            VARCHAR(255)     NOT NULL);''')

print("Table created successfully")

conn.commit()
```

```python
# insert data

cur = conn.cursor()

insert_query = """
    INSERT INTO assoc_rules_ips (ID, SUPPORT, IP)
    VALUES (%s, %s, %s)
"""

for idx, row in frequent_itemsets.iterrows():
    cur.execute(
        insert_query,
        (int(idx), float(row["support"]), str(list(row["itemsets"])))
    )

conn.commit()
print("Records created successfully")
conn.close()
```
