```python
import pandas as pd

df = pd.read_json("../../datas/nikto/detected.json", lines=True)

df.head(5)
```

```python
# get the paths with params
# filtered_paths = filtered_df["path"]

# get the paths without params
df["path"] = df["path"].apply(lambda x: x.split('?', 1)[0])

df.head(5)
```

```python
paths_matrix = []
tmp = []
sess_uuids_paths = {
  # "{sess_uuid}": []
}
sess_uuids = []

for _, row in df.iterrows():
  sess_uuid = row.get("cookies", {}).get("sess_uuid") or row.get("response_msg", {}).get("response", {}).get("message", {}).get("sess_uuid")
  path = row["path"]

  if sess_uuid in sess_uuids:
    sess_uuids_paths[sess_uuid].append(path)
  else:
    sess_uuids_paths[sess_uuid] = [path]
    sess_uuids.append(sess_uuid)

paths_matrix = list(sess_uuids_paths.values())
# for paths in sess_uuids_paths.values():
#   paths_matrix.append(paths)

sess_uuids_paths
# paths_matrix
# print(len(paths_matrix), len(sess_uuids_paths.keys()))
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(paths_matrix).transform(paths_matrix)
df = pd.DataFrame(te_ary, columns=te.columns_)

df
```

```python
from mlxtend.frequent_patterns import apriori

freq_itemsets = apriori(df, use_colnames=True, min_support=0.0015)

freq_itemsets
```

```python
freq_itemsets['length'] = freq_itemsets['itemsets'].apply(lambda x: len(x))
freq_itemsets
```
