```python
import pandas as pd

df = pd.read_json("../../datas/zap/detected.json", lines=True)

df.head(5)
```

```python
sess_uuids_matrix = []
tmp = []
sess_uuids_uuids = {
  # "{sess_uuid}": []
}
uuids = []

for _, row in df.iterrows():
  uuid = row["uuid"]
  sess_uuid = row.get("cookies", {}).get("sess_uuid") or row.get("response_msg", {}).get("response", {}).get("message", {}).get("sess_uuid")

  if uuid in uuids:
    sess_uuids_uuids[uuid].append(sess_uuid)
  else:
    sess_uuids_uuids[uuid] = [sess_uuid]
    uuids.append(uuid)

sess_uuids_matrix = list(sess_uuids_uuids.values())
# for paths in sess_uuids_uuids.values():
#   sess_uuids_matrix.append(paths)

sess_uuids_uuids
# sess_uuids_matrix
# print(len(sess_uuids_matrix), len(sess_uuids_uuids.keys()))
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth

te = TransactionEncoder()
te_ary = te.fit(sess_uuids_matrix).transform(sess_uuids_matrix)
df = pd.DataFrame(te_ary, columns=te.columns_)

df
```

```python
# frequent_itemsets = fpgrowth(df, min_support=0.3, use_colnames=True)
frequent_itemsets = fpgrowth(df, min_support=0.5, use_colnames=True)
### alternatively:
#frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
#frequent_itemsets = fpmax(df, min_support=0.6, use_colnames=True)

frequent_itemsets
```

```python
frequent_itemsets["length"] = frequent_itemsets["itemsets"].apply(lambda x: len(x))
### alternatively:
#frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
#frequent_itemsets = fpmax(df, min_support=0.6, use_colnames=True)

frequent_itemsets
```

```python
large_frequent_itemsets = frequent_itemsets[frequent_itemsets["length"] >= 3]
large_frequent_itemsets
```

```python
import psycopg2

conn = psycopg2.connect(database="web_honeypot_zap", user = "postgres", password = "admin", host = "127.0.0.1", port = "5432")

print("Opened database successfully")
```

```python
# create table
cur = conn.cursor()
cur.execute('''CREATE TABLE assoc_rules_sess_uuids (
            ID INT PRIMARY KEY     NOT NULL,
            SUPPORT           REAL    NOT NULL,
            SESS_UUID            TEXT     NOT NULL);''')

print("Table created successfully")

conn.commit()
```

```python
# insert data

cur = conn.cursor()

insert_query = """
    INSERT INTO assoc_rules_sess_uuids (ID, SUPPORT, SESS_UUID)
    VALUES (%s, %s, %s)
"""

for idx, row in frequent_itemsets.iterrows():
    if idx == 1000:
        break
    cur.execute(
        insert_query,
        (int(idx), float(row["support"]), str(list(row["itemsets"])))
    )

conn.commit()
print("Records created successfully")
```
