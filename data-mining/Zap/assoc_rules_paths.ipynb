```python
import pandas as pd

df = pd.read_json("../../datas/zap/detected.json", lines=True)

df.head(5)
```

```python
# get the paths with params
# filtered_paths = filtered_df["path"]

# get the paths without params
df["path"] = df["path"].apply(lambda x: x.split('?', 1)[0])

df.head(5)
```

```python
paths_matrix = []
tmp = []
sess_uuids_paths = {
  # "{sess_uuid}": []
}
sess_uuids = []

for _, row in df.iterrows():
  sess_uuid = row.get("cookies", {}).get("sess_uuid") or row.get("response_msg", {}).get("response", {}).get("message", {}).get("sess_uuid")
  path = row["path"]

  if sess_uuid in sess_uuids:
    sess_uuids_paths[sess_uuid].append(path)
  else:
    sess_uuids_paths[sess_uuid] = [path]
    sess_uuids.append(sess_uuid)

paths_matrix = list(sess_uuids_paths.values())
# for paths in sess_uuids_paths.values():
#   paths_matrix.append(paths)

sess_uuids_paths
# paths_matrix
# print(len(paths_matrix), len(sess_uuids_paths.keys()))
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth

te = TransactionEncoder()
te_ary = te.fit(paths_matrix).transform(paths_matrix)
df = pd.DataFrame(te_ary, columns=te.columns_)

df
```

```python
frequent_itemsets = fpgrowth(df, min_support=0.5, use_colnames=True)
### alternatively:
#frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
#frequent_itemsets = fpmax(df, min_support=0.6, use_colnames=True)

frequent_itemsets
```

```python
frequent_itemsets["length"] = frequent_itemsets["itemsets"].apply(lambda x: len(x))
### alternatively:
#frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
#frequent_itemsets = fpmax(df, min_support=0.6, use_colnames=True)

frequent_itemsets
```

```python
large_frequent_itemsets = frequent_itemsets[frequent_itemsets["length"] >= 3]
large_frequent_itemsets
```

```python
import psycopg2

conn = psycopg2.connect(database="web_honeypot_zap", user = "postgres", password = "admin", host = "127.0.0.1", port = "5432")

print("Opened database successfully")
```

```python
# create table
cur = conn.cursor()
cur.execute('''CREATE TABLE assoc_rules_paths (
            ID INT PRIMARY KEY     NOT NULL,
            SUPPORT           REAL    NOT NULL,
            PATH            TEXT     NOT NULL);''')

print("Table created successfully")

conn.commit()
```

```python
# insert data

cur = conn.cursor()

insert_query = """
    INSERT INTO assoc_rules_paths (ID, SUPPORT, PATH)
    VALUES (%s, %s, %s)
"""

for idx, row in frequent_itemsets.iterrows():
    cur.execute(
        insert_query,
        (int(idx), float(row["support"]), str(list(row["itemsets"])))
    )

conn.commit()
print("Records created successfully")
```
